```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr, arr.dtype)
```

```python
arr = np.array([1,2,3.14,5])
print(arr, arr.dtype)
```

```python
arr = np.array([1,2,3.14,5,"hello"])
print(arr, arr.dtype)
```

```python
#complex no.
arr = np.array([3+5j])
print(arr,arr.dtype)
```

```python
arr1 = np.array([3+5j])
arr2 = np.array([2+3j])
print(arr1 + arr2)
```

```python
arr = np.array(["prime", {1, 2, 3}, 3.14])
print(arr, arr.dtype)
```
