```python
import numpy as np
```

```python
#Mathematical fnx

#aggregation

arr = np.array( [1, 2, 3, 4, 5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))

print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))
print(np.var(arr))
```

```python
#power function
arr = np.array( [1, 2, 3, 4, 5])
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
#log
print(np.log(arr))
print(np.log10(arr))
print(np.log2(arr))
print(np.exp(arr))
```

```python
#rounding
print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))
```

```python
arr = np.array([1, 2, -8, 3, 4, -5, 2, -5])

print(np.unique(arr))
print(np.sort(arr))

print(np.abs(arr))
```
