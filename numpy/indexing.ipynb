```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr[0])
```

```python
#for 2d array
arr = np.array([[1,2,3],[4,5,6]])
print(arr[1][1])
```

```python
#fancy indexing
arr = np.array([1,2,3,4,5])
idx = [0,1,3]
print(arr[idx])
```

```python
#boolean indexing
arr = np.array([1,2,3,4,5])
print(arr[arr>3])
print(arr[arr%2==0])
```
