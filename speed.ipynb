```python
import numpy as np
import time
```

```python
size = 1_000_000
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")
array = np.array(py_list)
start = time.time()
sq_array = array**2
end = time.time()
print(f"numpy array time = {end-start} seconds")
```
