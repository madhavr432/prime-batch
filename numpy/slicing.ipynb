```python
# copy vs view
nums = [1,2,3,4,5]
sub_list = nums[1:3]
print(sub_list)
sub_list[0] = 200
print(sub_list)
print(nums)
arr = np.array([1,2,3,4,5])
sub_arr = arr[1:3]
print(sub_arr)
sub_arr[0] = 200
print(sub_arr)
print(arr)
arr1 = np.array([1,2,3,4,5])
sub_arr1 = arr1[1:3].copy()
print(sub_arr1)
sub_arr1[0] = 200
print(sub_arr1)
print(arr1)
```
